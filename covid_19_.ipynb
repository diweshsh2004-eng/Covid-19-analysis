```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#
```

```python
df= pd.read_csv("/content/Covid_19_Clean_Complete (2).csv")
```

```python
df.rename(columns={"Province/State" : "State"  , "Country/Region": "Country" } , inplace = True)
```

```python
df
```

```python
confirmed_cases = df.groupby (["Date"]) ["Confirmed"].sum().reset_index()
confirmed_cases
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(data=confirmed_cases, x="Date", y="Confirmed", color="red")
plt.xticks(rotation=90)

plt. show
```

```python
death_cases = df.groupby(["Date"])["Deaths"].sum().reset_index()
death_cases
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(data=death_cases, x="Date", y="Deaths", color="red")
plt.xticks(rotation=90)
```

```python
data_countries = df.groupby("Country")[["Confirmed","Active","Recovered","Deaths"]].sum().reset_index()
data_countries
```

```python
# Top 3 countries
top3_confirmed = data_countries.sort_values(by="Confirmed",ascending=False).head(3)
top3_confirmed
```

```python
top3_deaths = data_countries.sort_values(by="Deaths",ascending=False).head(3)
top3_deaths
```

```python
sns.barplot(data=top3_deaths, x="Country", y="Deaths", color="red")
plt.show
```

```python
from prophet import Prophet
```

```python
#Forecasting for the confirmed caes
```

```python
confirmed_cases
```

```python
prophet_confirmed = confirmed_cases.rename(columns={"Date":"ds","Confirmed":"y"})
prophet_confirmed
```

```python
model = Prophet()
model.fit(prophet_confirmed)
```

```python
future = model.make_future_dataframe(periods=7) #for future 7 days
future
```

```python
forecast = model.predict(future)
forecast
```

```python
forecast[["yhat_lower","yhat_upper", "yhat"]]
```

```python
model.plot(forecast)
plt.show()
```
